# Introduction to Graph theory

## Basic Graphs using networkx

```python
# First load some basic packages
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
# To create graph
G=nx.Graph()
#add edges
G.add_node(1)
# add more nodes
G.add_node(2)
G.add_node(3)
G.add_node(4)
G.add_node(5)

# To check total number of existing nodes
G.nodes()
```

```python
# Add edges
G.add_edge(1,2)
G.add_edge(2,3)
G.add_edge(3,4)
G.add_edge(4,5)

# To check total numberof edges
G.edges()
```

```python
# To visualise
nx.draw(G)
```

```python
# to see labels
nx.draw(G,with_labels=1)
```

```python
# to change colors of the graphs
nx.draw(G,with_labels=1,node_color='r')
```

```python
# add one more edge
G.add_edge(5,6)
```

```python
nx.draw(G,with_labels=1)
```

## Degree Sequence of a Graph or not

The *degree* of a vertex on a simple graph is the number of edges connected to that vertex. The *degree sequence* is the sequence of these degree values, arranged from the largest value to smallest.

If $G$ is a graph on $n$ vertices $v_1, v_2,\cdots, v_n$ with degrees $d_1, d_2, \cdots, d_n$ respectively, then the $n$-tuple $(d_1,d_2,\cdots,d_n)$ is called the *degree sequence* of $G$.

A sequence of non-negative integers $S=(d_1,d_2,\cdots,d_n)$ is said to be graphical, if there exists a simple graph $G$ with $n$ vertices $v_1, v_2, \cdots, v_n$ such that $deg(v_i)=d_i$, for $i=1,2,\cdots,n$.

```python
P=(2,2,2)
nx.is_graphical(P)
```

```python
nx.is_graphical(P, method='hh')
```

```python
nx.is_graphical(P,method='eg')
```
